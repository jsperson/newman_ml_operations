# Assignment 1.2 — Wide to Thin Reshape

Here's the structural problem the data has. Every feature you'll build in this course — recent win rates, LP trajectories, champion pools, role consistency, tilt signals — lives at the player level. But the data arrives at the game level, one row per game, with each of the 10 players encoded as a set of prefix columns (b1_, b2_, ..., r5_).
You can't engineer player-level features from a player-level history if every player is smeared across ten columns in one row. The first piece of infrastructure this course needs is a reshape: from one row per game to ten rows per game, one per player.

## pd.melt vs. pd.wide_to_long
pd.wide_to_long() works best when columns share a stub name and differ by a suffix, such as score_1, score_2, score_3. This dataset uses prefixes like b1_puuid, r5_championId, so using wide_to_long() would require renaming columns first. pd.melt() is more flexible because it can turn arbitrary columns into key/value rows, but for this dataset it creates a very large intermediate table and then requires another pivot. I ultimately used an explicit slot-loop approach because it preserves column structure directly and avoids the memory overhead of melt.

I'm using the Hermes agent framework backed by gpt-5.5 for AI assistance. After some discussion our conclusion is that pd.melt() is the better option. pd.wide_to_long is a good function but it works more naturally when the repeated-observation identifier appears as a suffix after a shared variable stub. Our columns have the identifier at the end of the prefix. We could rename the columns but that's an extra step and we could just use melt. This seems reasonable and there is very little overhead to implement the more complex melt syntax if AI assisted.

## bwin vs. win
In the original wide data we are provided with bwin which indicates if bwin won the respective match. This is appropriate for the wide structure because each row contains attributes and the result for both blue and red teams. An rwin column would work as well but having both would be redundant as one is the exact inverse of the other.

When we reverse pivot the data and have one record per player we could have retained bwin but then we'd have to carry computational logic for calculating whether a player won based on their team membership throughout the project. It's logical that we'd want to eliminate this complexity and just have a "win" value. To do this we take the bwin value as is for blue team members and inverse (or use 1-bwin) for the red team.

## Column changes
For the narrow format we will (by definition) not have as many columns. The bx and rx columns are reverse pivoted into one player attribute set per record. bwin is transformed into win using rules based on the player team.

I also considered moving the match level attributes to a second target table. This would have eliminated a lot of data duplication through normalization. In the end I decided that the simplicity of a single "wide" narrow table was preferrable to having to carry join logic and compute overhead through the rest of the project.

## Reshape implementation
I originally had this coded up using melt but it wasn't great. Due to the number of records and columns memory consumption blew up colab so I had to add a loop and do the pivot in chunks. This overhead more than exceeded any efficiencies from using melt so I changed strategies.

For my implementation I chose the none of the above option. I experimented with melt but it required a melt which created a key/value pair dataframe (a huge one) that was then reverse pivoted back into the desired shape. This was slow. My implementation below uses looping to work through the columns and create recods of the right shape. These records are then appended to the dataframe.

There are some additional fields that are somewhat duplicative. In particular bwin/win and side_code/side_slot/player_slot don't add data but they might be useful for validation or analysis. I'll leave them in for now.

Note that I have a bit of a weird return structure. I needed the match_cols list in order to validate the values in those columns later.

```python
def make_player_match_df(full_df):
    """Convert full_df from one row per match to one row per player per match.

    The source dataframe has one row per match and player columns named like:
        b1_puuid, b1_teamPosition, ..., r5_mpoints

    The returned dataframe has ten rows per match: one row for each blue/red
    player slot, with all match-level attributes retained.
    """
    player_col_pattern = re.compile(r"^[br][1-5]_")
    player_cols = [
        col for col in full_df.columns
        if player_col_pattern.match(col)
    ]
    match_cols = [
        col for col in full_df.columns
        if col not in player_cols
    ]

    if not player_cols:
        raise ValueError("No player columns found. Expected names like b1_puuid or r5_championId.")

    slot_specs = [
        ("b", "blue", 1),
        ("b", "blue", 2),
        ("b", "blue", 3),
        ("b", "blue", 4),
        ("b", "blue", 5),
        ("r", "red", 1),
        ("r", "red", 2),
        ("r", "red", 3),
        ("r", "red", 4),
        ("r", "red", 5),
    ]

    player_frames = []
    for side_code, team, side_slot in slot_specs:
        player_slot = f"{side_code}{side_slot}"
        prefix = f"{player_slot}_"
        slot_cols = [
            col for col in player_cols
            if col.startswith(prefix)
        ]

        if not slot_cols:
            raise ValueError(f"No columns found for player slot {player_slot}.")

        slot_df = full_df[match_cols + slot_cols].copy()
        rename_map = {
            col: col.removeprefix(prefix)
            for col in slot_cols
        }
        slot_df = slot_df.rename(columns=rename_map)

        insert_at = len(match_cols)
        slot_df.insert(insert_at, "team", team)
        slot_df.insert(insert_at + 1, "side_code", side_code)
        slot_df.insert(insert_at + 2, "side_slot", side_slot)
        slot_df.insert(insert_at + 3, "player_slot", player_slot)

        player_frames.append(slot_df)

    player_match_df = pd.concat(player_frames, ignore_index=True)

    # bwin is 1 when blue won; win is 1 when this player's team won.
    team_is_blue = player_match_df["team"].eq("blue").to_numpy()
    bwin = player_match_df["bwin"].to_numpy(dtype=np.int8, copy=False)

    player_match_df["win"] = np.where(team_is_blue, bwin, 1 - bwin).astype(np.int8)

    # Convert date strings to pandas datetime for efficient date operations.
    if "gameCreationDate" in player_match_df.columns:
        player_match_df["gameCreationDate"] = pd.to_datetime(
            player_match_df["gameCreationDate"],
            format="%m/%d/%Y"
        )

    # Convert whole-number columns to smaller integer dtypes after the reshape.
    # Use nullable integer dtypes for source columns because they may contain missing values.
    integer_dtype_map = {
        "leaguePoints": "Int16",
        "championId": "Int16",
        "mpoints": "Int32",
        "side_slot": "int8",
        "bwin": "int8",
    }

    for col, dtype in integer_dtype_map.items():
        if col in player_match_df.columns:
            player_match_df[col] = player_match_df[col].astype(dtype)

    return player_match_df, match_cols
```

```python
# Run after full_df is loaded:
player_match_df, match_cols = make_player_match_df(full_df)
print(f"full_df shape: {full_df.shape}")
print(f"player_match_df shape: {player_match_df.shape}")
player_match_df.head(10)
```

## Reshape validation

```python
# Validation checks for make_player_match_df
expected_slots = {"b1", "b2", "b3", "b4", "b5", "r1", "r2", "r3", "r4", "r5"}

# Test 1: each match should become exactly ten player rows.
assert len(player_match_df) == len(full_df) * 10

# Test 2: no game should lose or gain players during the reshape.
players_per_game = player_match_df.groupby("gameId").size()
assert players_per_game.eq(10).all()

# Test 3: each game should have exactly one row for each expected player slot.
slots_per_game = player_match_df.groupby("gameId")["player_slot"].agg(set)
assert slots_per_game.apply(lambda slots: slots == expected_slots).all()

# Test 4: gameId + player_slot should uniquely identify a player row.
assert not player_match_df.duplicated(["gameId", "player_slot"]).any()

# Test 5: each game should have five blue rows and five red rows.
team_counts = (
    player_match_df
    .groupby(["gameId", "team"])
    .size()
    .unstack(fill_value=0)
    .reindex(columns=["blue", "red"], fill_value=0)
)
assert team_counts["blue"].eq(5).all()
assert team_counts["red"].eq(5).all()

# Test 6: blue player outcomes should equal bwin; red outcomes should be the inverse.
if "bwin" in player_match_df.columns:
    blue_rows = player_match_df["team"].eq("blue")
    red_rows = player_match_df["team"].eq("red")
    assert (player_match_df.loc[blue_rows, "win"] == player_match_df.loc[blue_rows, "bwin"]).all()
    assert (player_match_df.loc[red_rows, "win"] == 1 - player_match_df.loc[red_rows, "bwin"]).all()

# Test 7: whole-number columns should use compact integer dtypes.
expected_dtypes = {
    "leaguePoints": "Int16",
    "championId": "Int16",
    "mpoints": "Int32",
    "side_slot": "int8",
    "bwin": "int8",
    "win": "int8",
    "gameCreationDate": "datetime64[ns]"
}
actual_dtypes = player_match_df[list(expected_dtypes)].dtypes.astype(str).to_dict()
assert actual_dtypes == expected_dtypes, actual_dtypes

# Test 8: validate that the metadata came over properly (unique for each source record)
metadata_cols = [
    col for col in match_cols
    if col in player_match_df.columns
]
metadata_nunique = (
    player_match_df
    .groupby("source_row_id")[metadata_cols]
    .nunique(dropna=False)
)
assert not metadata_nunique.gt(1).any().any()

print("All reshape validation checks passed.")
```

## Code to build data dictionary

```python
from IPython.display import Markdown, display
from pandas.api.types import is_integer_dtype, is_float_dtype, is_datetime64_any_dtype

# Build a data dictionary for the thin/player-level table.
# Distinct counts include missing values, if present.

descriptions = {
    "gameId": "Unique identifier for the League of Legends match.",
    "gameCreation": "Raw match creation timestamp from the source data.",
    "gameCreationDate": "Calendar date when the match was created, converted to pandas datetime.",
    "gameDuration": "Length of the match in seconds.",
    "patch": "League of Legends patch/version identifier for the match.",
    "tier": "Ranked tier for the match/player population.",
    "average_leaguePoints": "Average League Points for the match in the source data.",
    "bwin": "Binary source outcome: 1 if the blue team won, 0 if the red team won.",
    "championId": "Riot champion identifier for this player row.",
    "leaguePoints": "League Points for this player row.",
    "mpoints": "Champion mastery points for this player/champion row.",
    "puuid": "Anonymized Riot player UUID for this player row.",
    "teamPosition": "Player role/lane position for this row.",
    "team": "Team color for this player row: blue or red.",
    "side_code": "Short team-side code from the original wide column prefix: b or r.",
    "side_slot": "Player slot number within a team, 1 through 5.",
    "player_slot": "Original wide-table player slot identifier, such as b1 or r5.",
    "win": "Binary player-level outcome: 1 if this player's team won, 0 otherwise.",
}

categorical_columns = {"tier", "teamPosition", "team", "side_code", "side_slot", "player_slot", "bwin", "win", "patch"}
string_columns = {"puuid"}
low_cardinality_limit = 25


def md_escape(value):
    text = str(value)
    return text.replace("|", "\\|").replace("\n", " ")


def semantic_type(series):
    name = series.name
    if name in categorical_columns:
        return "categorical"
    if name in string_columns:
        return "string"
    if is_datetime64_any_dtype(series):
        return "datetime"
    if is_integer_dtype(series):
        return "int"
    if is_float_dtype(series):
        return "float"
    return "string"


def range_or_values(series, dtype_label, distinct_count):
    if dtype_label in {"int", "float"}:
        return f"min={md_escape(series.min())}, max={md_escape(series.max())}"

    if dtype_label == "datetime":
        return f"min={md_escape(series.min())}, max={md_escape(series.max())}"

    values = series.dropna().drop_duplicates()
    try:
        values = values.sort_values()
    except Exception:
        pass

    if distinct_count <= low_cardinality_limit:
        return ", ".join(md_escape(v) for v in values.tolist())

    return f"many ({distinct_count} distinct)"

rows = []
for col in player_match_df.columns:
    series = player_match_df[col]
    dtype_label = semantic_type(series)
    distinct_count = int(series.nunique(dropna=False))
    rows.append({
        "Column name": col,
        "Data type": dtype_label,
        "Description": descriptions.get(col, "Column carried forward from the source table."),
        "Range or values": range_or_values(series, dtype_label, distinct_count),
        "Distinct count": distinct_count,
    })

data_dictionary_df = pd.DataFrame(rows)
data_dictionary_markdown = "## Data dictionary for `player_match_df`\n\n" + data_dictionary_df.to_markdown(index=False)
display(Markdown(data_dictionary_markdown))
```

## Why change to a narrow format?
We're not done with this project! I anticipate that we'll do quite a bit of feature engineering with this dataset. For a given game a particular user could be in any one of ten positions. With such a wide format any feature that depends on user level attributes would need to search through or aggregate across all ten columns of each attribute. That would be a pain to code - effectively you'd be doing a reverse pivot with every pass through the data.

A narrow format moves the user attributes to a single set of columns. Aggregations and grouping can be done vertically the way databases and these types of functions are designed to work.
